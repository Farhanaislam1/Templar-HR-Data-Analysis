```python
import pandas as pd

df = pd.read_csv("TemplarHR.csv")
```

**1. Compare employee salary by gender for the company in each department, factoring in years of service:**

```python
import seaborn as sns
from IPython.display import display
import matplotlib.pyplot as plt

# Group the data by 'Department'
grouped_df = df.groupby('Department')

# Iterate over each group and create a visualization
for department, data in grouped_df:
    sns.scatterplot(data=data, x='YearsAtCompany', y='Annual Rate', hue='Gender', palette={'Male': 'blue', 'Female': 'magenta'})
    plt.title(f"Annual Rate vs Years at Company - Department: {department}")
    plt.xlabel('Years at Company')
    plt.ylabel('Annual Rate')
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.show()

    # Display descriptive table for Male gender
    print('Descriptive table for Male gender')
    male_table = data[data['Gender'] == 'Male'][['Job level', 'Annual Rate', 'YearsAtCompany']].describe()
    display(male_table)

    # Display descriptive table for Female gender
    print('Descriptive table for Female gender')
    female_table = data[data['Gender'] == 'Female'][['Job level', 'Annual Rate', 'YearsAtCompany']].describe()
    display(female_table)  
    
    # % Variance of Male vs Female descriptive tables
    print("% Variance of Male vs Female descriptive tables")
    gender_gap = male_table/female_table - 1
    display(gender_gap)
```

**2. Compare salaries by gender and minority status for job types level 2 or higher:** filter the data for job types level 2 or higher and compare salaries by gender and minority status.

```python
# Filter for job types level 2 or higher
higher_level_jobs = data[data['Job level'] >= 2]

# Group by gender and minority status, calculate average salary
avg_salary_by_gender_minority = higher_level_jobs.groupby(['Gender', 'Visible_Minority'])['Annual Rate'].mean().reset_index()
avg_salary_by_gender_minority
```

```python
# Calculate median and percentiles for salary by gender and minority status
median_salary_by_gender_minority = higher_level_jobs.groupby(['Gender', 'Visible_Minority'])['Annual Rate'].median().reset_index()
percentile25_salary_by_gender_minority = higher_level_jobs.groupby(['Gender', 'Visible_Minority'])['Annual Rate'].quantile(0.25).reset_index()
percentile75_salary_by_gender_minority = higher_level_jobs.groupby(['Gender', 'Visible_Minority'])['Annual Rate'].quantile(0.75).reset_index()

# Rename the column for 25th and 75th percentile
percentile25_salary_by_gender_minority.rename(columns={'Annual Rate': 'Annual Rate_25th_percentile'}, inplace=True)
percentile75_salary_by_gender_minority.rename(columns={'Annual Rate': 'Annual Rate_75th_percentile'}, inplace=True)

# Merge the dataframes
summary_salary_by_gender_minority = avg_salary_by_gender_minority.merge(median_salary_by_gender_minority, on=['Gender', 'Visible_Minority'], suffixes=('_mean', '_median'))
summary_salary_by_gender_minority = summary_salary_by_gender_minority.merge(percentile25_salary_by_gender_minority, on=['Gender', 'Visible_Minority'])
summary_salary_by_gender_minority = summary_salary_by_gender_minority.merge(percentile75_salary_by_gender_minority, on=['Gender', 'Visible_Minority'], suffixes=('', '_75th_percentile'))

summary_salary_by_gender_minority
```

```python
# Visualize the results

# Set the style of seaborn
sns.set(style="whitegrid")

# Create the box plot
plt.figure(figsize=(10, 8))
sns.boxplot(data=higher_level_jobs, x='Gender', y='Annual Rate', hue='Visible_Minority')
plt.title('Salary Distribution by Gender and Minority Status')
plt.xlabel('Gender')
plt.ylabel('Annual Salary')
plt.legend(title='Visible Minority')
plt.show()
```

**Insights:** Overall, the analysis reveals disparities in salaries based on gender and minority status for job types level 2 or higher. While some differences exist, further exploration is needed to understand the intersectional dynamics and potential contributing factors.

* **Gender Comparison:** On average, male employees earn a higher annual salary (`$89,857.09`) compared to female employees (`$83,155.57`) for job types level 2 or higher, indicating a gender pay gap of approximately `$6,701.52`. Additionally, the analysis of salary percentiles reveals that male employees tend to have higher positions in the salary distribution, with the 75th percentile for males being notably higher than that for females. This suggests that while the mean salary difference is significant, the disparity is even more pronounced at higher salary levels.

* **Minority Status Comparison:** Employees who do not identify as a visible minority earn a higher average salary (`$89,857.09`) compared to those who identify as a visible minority (`$85,261.66`) for job types level 2 or higher, suggesting a disparity in average salary based on minority status of approximately `$4,595.43`.Further analysis of salary percentiles reveals that non-visible minority employees also tend to occupy higher percentiles in the salary distribution compared to visible minority employees. This indicates a potential discrepancy in the distribution of salaries between these two groups, with non-visible minority employees having a higher representation at higher salary levels.

* **Intersectional Analysis:** Further investigation is warranted to understand the intersectionality of gender and minority status in salary disparities for job types level 2 or higher. Analyzing the combined effects of gender and minority status may reveal nuanced insights into pay discrepancies within the organization, particularly at different percentiles of the salary distribution.

These insights underscore the importance of addressing disparities in salaries based on gender and minority status and implementing strategies to promote pay equity and diversity within the organization. Further analysis and exploration of potential contributing factors can inform targeted interventions aimed at fostering a more inclusive and equitable workplace environment.

**3. Find if common job titles have differences in pay in different departments:** identify common job titles across departments and compare their salaries.

```python
# Identify common job titles
common_job_titles = ['Analyst', "Clerk", "Director", "Manager", "Tech"]

# Filter for common job titles
common_jobs_data = df[df['Job Title'].isin(common_job_titles)]

# Group by job title and department, calculate average salary
avg_salary_by_job_department = common_jobs_data.groupby(['Job Title', 'Department'])['Annual Rate'].mean().reset_index()
avg_salary_by_job_department
```

```python
# Calculate median and percentiles for salary by job title and department
median_salary_by_job_department = common_jobs_data.groupby(['Department', 'Job Title'])['Annual Rate'].median().reset_index()
percentile25_salary_by_job_department = common_jobs_data.groupby(['Department', 'Job Title'])['Annual Rate'].quantile(0.25).reset_index()
percentile75_salary_by_job_department = common_jobs_data.groupby(['Department', 'Job Title'])['Annual Rate'].quantile(0.75).reset_index()

# Rename the column for 25th and 75th percentile
percentile25_salary_by_job_department.rename(columns={'Annual Rate': 'Annual Rate_25th_percentile'}, inplace=True)
percentile75_salary_by_job_department.rename(columns={'Annual Rate': 'Annual Rate_75th_percentile'}, inplace=True)

# Merge the dataframes
summary_salary_by_job_department = avg_salary_by_job_department.merge(median_salary_by_job_department, on=['Department', 'Job Title'], suffixes=('_mean', '_median'))
summary_salary_by_job_department = summary_salary_by_job_department.merge(percentile25_salary_by_job_department, on=['Department', 'Job Title'], suffixes=('', '_25th_percentile'))
summary_salary_by_job_department = summary_salary_by_job_department.merge(percentile75_salary_by_job_department, on=['Department', 'Job Title'], suffixes=('', '_75th_percentile'))
summary_salary_by_job_department
```

```python
# Visualize the results

plt.figure(figsize=(12, 6))
sns.barplot(data=summary_salary_by_job_department, x='Department', y='Annual Rate_mean', hue='Job Title')
plt.xticks(rotation=45)
plt.title('Average Salary by Department and Job Title')
plt.xlabel('Department')
plt.ylabel('Average Annual Salary')
plt.legend(title='Job Title', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# # Visualize the results: create separate box plots for each job title

plt.figure(figsize=(12, 8))

for i, job_title in enumerate(common_job_titles):
    plt.subplot(2, 3, i+1)
    sns.boxplot(data=common_jobs_data[common_jobs_data['Job Title'] == job_title], x='Department', y='Annual Rate')
    plt.title(job_title)
    plt.xlabel('Department')
    plt.ylabel('Annual Salary')
    plt.xticks(rotation=45)  # Rotate x-axis labels

plt.tight_layout()
plt.show()
```

**Insights:** The analysis examines differences in salaries for common job titles across various departments, shedding light on potential disparities in pay within the organization.
* **Analyst Positions:** Analysts in the Finance department have an average annual salary of `$92,930.36`, with a median salary of `$92,090.50`. The salary distribution for Analysts in Finance is relatively tight, with the 25th percentile at `$85,249.75` and the 75th percentile at `$97,674.00`. In contrast, Analysts in IT have a lower average salary of `$69,673.82` but a wider salary range, with the 25th percentile at `$59,439.00` and the 75th percentile at `$91,434.00`. Similar patterns are observed for Analysts in Logistics, Operations, and Sales & Marketing, indicating variations in pay and salary distribution across departments for this role.
* **Clerk Positions:** Clerks in the Finance department have an average annual salary of `$41,655.58`, with a median salary of `$40,782.00`. The salary distribution for Clerks in Finance is relatively balanced, with the 25th percentile at `$31,635.00` and the 75th percentile at `$47,729.00`. In comparison, Clerks in IT, Logistics, and Operations exhibit similar trends in average salary and median salary, but with varying salary ranges across percentiles.
* **Manager Positions:** Managers in the Finance department have an average annual salary of `$116,043.25`, with a median salary of `$115,689.50`. The salary distribution for Managers in Finance is relatively tight, with the 25th percentile at `$110,714.50` and the 75th percentile at `$121,372.25`. Conversely, Managers in IT, Logistics, Operations, and Sales & Marketing demonstrate wider salary ranges across percentiles, indicating variations in pay distribution across departments for Managerial roles.

These insights provide valuable information about the salary discrepancies for common job titles across different departments within the organization, considering both average salary and salary distribution across percentiles. Addressing these variations in pay can contribute to promoting fairness and equity in compensation practices. Further analysis and exploration may be necessary to identify underlying factors contributing to these differences and inform targeted interventions aimed at fostering a more equitable workplace environment.

**4. Assess if management teams reflect the demographics of the workforce they manage:** 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("TemplarHR.csv")
```

```python
df['Job Category'] = df['Job Title'].apply(lambda x: 'Management' if x in ['Manager', 'Director'] else 'Non-Management')
```

```python
summary_table = df.pivot_table(values='employee_ID', index=['Job Category'], columns=['Department'], aggfunc='count', margins=True, margins_name='Total')
summary_table
```

```python
# Calculating the percentage of employees in each department by job category
subtotal_values = summary_table.loc['Total']
summary_table_percent = summary_table.div(subtotal_values, axis=1) * 100
summary_table_percent
```

```python
transposed_table = summary_table_percent.transpose()
transposed_table.drop('Total', axis=1, inplace=True)
transposed_table
```

```python
transposed_table.drop('Total').plot(kind='barh', stacked=True, color=['#00E0C2', '#E4DFDB'], fontsize=10)

# Add labels inside the bars
for i, (index, row) in enumerate(transposed_table.drop('Total').iterrows()):
    for j, value in enumerate(row):
        if j == 0:  # 'Management' category
            plt.annotate(f'{value:.2f}%', (value / 2, i), va='center', ha='center', color='black', fontsize=6)
        else:  # 'Non-Management' category
            plt.annotate(f'{value:.2f}%', (row[0] + value / 2, i), va='center', ha='center', color='black', fontsize=6)


plt.legend(bbox_to_anchor=(0, -0.2), loc='upper left', ncol=2, fontsize=10)
plt.title('Management Distribution by Department', fontsize=12)
plt.show()
```

In terms of the Management demographics, the distribution of managers and directors across the departments revealed that the Finance department stands out from the rest of the company by having 19.1% of its team in management positions whereas the Operations department has the lowest proportion (5.21%). The three other departments have around 9-10 % in the management team.

```python
# Calculating the count of employees in each department by job category and gender
summary_gender = df.pivot_table(values='employee_ID', index=['Gender'], columns=['Department','Job Category'], aggfunc='count', margins=True, margins_name='Total')
summary_gender
```

```python
# Calculating the percentage of employees in each department by job categorya and gender
subtotal_values = summary_gender.loc['Total']
summary_gender_percent = summary_gender.div(subtotal_values, axis=1) * 100
summary_gender_percent
```

```python
transposed_gender = summary_gender_percent.transpose()
transposed_gender.drop('Total', axis=1, inplace=True)
transposed_gender
```

```python
transposed_gender.drop('Total').plot(kind='barh', stacked=True, color=['magenta', 'blue'], fontsize=10)

# Add labels inside the bars
for i, (index, row) in enumerate(transposed_gender.drop('Total').iterrows()):
    for j, value in enumerate(row):
        if j == 0:  # 'Female' category
            plt.annotate(f'{value:.2f}%', (value / 2, i), va='center', ha='center', color='white', fontsize=8)
        else:  # 'Male' category
            plt.annotate(f'{value:.2f}%', (row[0] + value / 2, i), va='center', ha='center', color='white', fontsize=8)

plt.legend(bbox_to_anchor=(0, -0.2), loc='upper left', ncol=2, fontsize=10)
plt.title('Management Distribution by Gender and Department', fontsize=12)
plt.show()
```

For deeper insights into the subject, a similar analysis was conducted by adding the gender topic to the equation.  The results show that Logistics and IT management reflect the demographics of their workforce, while Finance and Operations are the most concerning departments suggesting that male employees from those areas are more likely to be promoted to a manager or director position. The gender distribution in management roles in the Sales & Marketing team is relatively balanced but with room for improvement in terms of gender equality.

```python
# Pivot the data to get the count of each age group for each department
pivot_table = df.pivot_table(values='employee_ID', index=['Department', 'Job Category'], columns=['Visible_Minority'], aggfunc='count')

# Calculate the total count for each department
total_count = pivot_table.sum(axis=1)

# Calculate the percentage of each age group within each department
percentage_table = pivot_table.div(total_count, axis=0) * 100

# Plot the stacked bar chart
ax = percentage_table.plot(kind='barh', stacked=True, color=['#E4DFDB','#00E0C2'])

# Add labels inside the bars
for i, (index, row) in enumerate(percentage_table.iterrows()):
    cumsum = 0  # track the end of the last bar segment
    for j, value in enumerate(row):
        # calculate the mid point for the text
        mid_point = cumsum + (value / 2)
        # only add text if the segment is large enough
        if value > 5:
            ax.text(mid_point, i, f'{value:.1f}%', va='center', ha='center', color='black', fontsize=8)
        cumsum += value

# Add labels and title
plt.legend(bbox_to_anchor=(0.5, -0.2), loc='upper center', ncol=5, fontsize=10)
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Distribution of Minorities by Department (Stacked to 100%)')

# Show the plot
plt.show()
```

The provided data underscores notable disparities in the representation of visible minorities between management and non-management roles across various departments within the organization, as well as variations in minority representation among departments.

Overall Patterns:
Across all departments, the percentage of visible minorities in management roles consistently falls within the range of 10-15%, while in non-management roles, it varies more widely, ranging from 18-36%. This suggests a systemic underrepresentation of visible minorities in leadership positions compared to their presence in non-management roles across the organization.

Operations Department:
The Operations department exhibits the highest percentage of visible minorities in non-management roles, with 36% representation. However, this representation decreases to 10-15% in management roles, aligning with the overall pattern observed across departments.

Finance Department:
In the Finance department, the percentage of visible minorities in non-management roles is 18%, which is lower than the Operations department but still within the broader range observed across departments. Similarly, the representation of visible minorities in management roles in Finance falls within the 10-15% range, consistent with the overall pattern.

Sales & Marketing Department:
The Sales & Marketing department demonstrates a relatively high percentage of visible minorities in non-management roles, with 33.6% representation. However, there is a substantial drop in representation in management roles, where visible minorities account for only 14.3%. This disparity highlights challenges in the advancement of visible minorities into leadership positions within this department.

Logistics Department and IT Department:
Both the Logistics and IT departments also exhibit significant disparities in minority representation between non-management and management roles. While there is notable representation of visible minorities in non-management roles (25.7% and 28.4% respectively), this decreases to 12.5% and 15.4% in management roles, indicating barriers to advancement for visible minorities within these departments.

In summary, while there are variations in minority representation among departments in non-management roles, disparities persist in the progression of visible minorities into management positions across the organization. Addressing these disparities requires concerted efforts to promote diversity and inclusion, including targeted recruitment, leadership development programs, and initiatives to mitigate biases in hiring and promotion processes. By fostering an inclusive workplace culture and creating pathways for advancement, organizations can work towards ensuring equitable representation of visible minorities in leadership roles across all departments.
